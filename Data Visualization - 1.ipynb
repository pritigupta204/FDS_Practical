```python
import matplotlib.pyplot as plt
plt.figure(figsize = (12 , 7))
# categorical data 
countries =  ['USA', 'Brazil', ' Russia', 'spain', 'UK', 'India']
totalDeaths = [112596, 37312, 5971 , 27136, 40597, 7449]
plt.bar(countries,totalDeaths , width = 0.9, align = 'center' , color = 'cyan',edgecolor = 'red')

i = 1.0
j = 2000

for i in range(len(countries)):
    plt.annotate(totalDeaths[i], (-0.1+ i, totalDeaths[i] + j))
    
plt.legend(labels = [' Total Deaths'])

plt.title(" Bar plot representing the total deaths by top 6 countries")

plt.xlabel("Cuntries")
plt.ylabel("Deaths")


plt.savefig('1BarPlot,png')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = [15,10])

totalDeath = [113055, 37312, 5971, 7473, 33964]
totalRecovery = [773480, 325603, 230688, 129095, 166584]
activeCase = [1139953, 324643, 234444, 342222,353323]
country = ['USA', 'Bra','Ru','India','Italy']

X = np.arange(len(totalDeath))

plt.bar(X, totalDeath , color = 'black',width = 0.25)
plt.bar(X + 0.25, totalRecovery, color = 'g', width = 0.25)
plt.bar(X + 0.5, activeCase, color = 'r', width = 0.25)

plt.legend(['Toatal Deaths', 'Total Recovary', 'Active Cases'])
plt.xticks([i + 0.25 for i in range(5)], country)

plt.title("Bar plot representing the total deaths, total recovery cases and active cases country wise ")

plt.xlabel("Country")
plt.ylabel("Case")
plt.show()
```

Pie Chart

```python
from pandas import DataFrame
import matplotlib.pyplot as plt

Data = { 'Tasks' : [400, 300, 300],
       'Task type' : ['tp', 'to','tc']
       }
df = DataFrame(Data)

df.set_index('Task type', inplace = True)    #, inplace = True)

df.plot.pie(y = 'Tasks', figsize = (5,5), autopct = '%1.1f%%' , startangle = 90)
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

labels = ['bcs', 'ca','it','dsai']
sizes = [20,45,15,20]

explode = (0.1, 0.1, 0.1, 0.4)

fig, ax = plt.subplots()
ax.pie(sizes, 
      explode = explode,
      autopct  = '%1.1f%%' ,
      shadow = True,
    startangle = 90 )
ax.axis('equal')
```
